- 2017~2022년도 데이터

```python
df["연도"]= df["시점"].str.split(".").str[0].values
df["월"]= df["시점"].str.split(".").str[1].values

day_values = df['월'].str.split('/')
df['월'] = day_values.str[0]
df['일'] = day_values.str[1]
df.head()
```

```python
df =df.rename(columns={"데이터":"백만"})
```

```python
df.info()
```

```python
df.loc[df['일']=='4 p)',"일"] =4
df.head()
```

```python
df[["연도","월","일"]]= df[["연도","월","일"]].astype(int)
```

```python
# 분기를 월이라고 해버렸네요..^^
df.describe()
```

- 백만 데이터가 편차가 심해보여서 정규화가 필요하지 않을까 싶다..



```python
#@title 📌 Q2. pivot_table을 사용하여 국가(대륙)별 연도별 판매액의 합계를 분석해 주세요.
from pandas.core.reshape.pivot import pivot_table
df_pivot= pivot_table(data=df , values="백만" ,columns="연도", index="국가(대륙)별",aggfunc="sum")
df_pivot
```

```python
plt.figure(figsize=(10,12))
df_pivot.plot.barh()
```

```python
# 연도에 따른 국가의 판매액 흐름
df_pivot.transpose().plot()
```

```python
#@title 📌 Q3. groupby를 사용하여 2020년의 온라인 해외판매 상품군을 지역별 합계를 구해 분석해 주세요
df.head()
```

```python
df["상품군별"].value_counts()
```

```python
#@title 2020년도 데이터 추출
df_2020 = df.loc[df["연도"]==2020]
df_2020.head()
```

```python
df_2020_sales = df_2020.groupby(["국가(대륙)별","상품군별"])["백만"].sum().unstack()
df_2020_sales
```

```python
df_2020_sales.index
```

```python
display(df_2020_sales.loc[["미국","일본","중국","아세안(ASEAN)"]].plot.barh())
display(df_2020_sales.loc[['기타', '대양주','유럽연합(EU)','중남미', '중동']].plot.barh())
```

```python
#@title 📌 Q4. 주요 판매 국가와 상품군에 대해 2021년까지의 연도별 온라인 직접 판매액 합계를 시각화 해주세요
df_max_2021 = df.loc[df["연도"]<2022]
```

```python
con_country= (df["국가(대륙)별"]=="미국")|(df["국가(대륙)별"]=="중국")|(df["국가(대륙)별"]=="일본")
con_product = (df["상품군별"]=="의류 및 패션 관련상품")|(df["상품군별"]=="화장품")|(df["상품군별"]=="음반·비디오·악기")

df_max_2021=df_max_2021.loc[con_country&con_product]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(df_max_2021, col="국가(대륙)별", hue="상품군별")
g.map(sns.lineplot, "연도", "백만")
plt.figure(figsize=(12,15))
g.set_axis_labels("연도", "백만")

g.add_legend(title="상품군별")

plt.show()
```
