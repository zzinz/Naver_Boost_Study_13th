#### **üåù FacetGridÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞**

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/NHIS_OPEN_GJ_2017.CSV.zip", encoding="cp949")

# Îß§Ìïë ÎîïÏÖîÎÑàÎ¶¨
smoke = {1 : "Ìù°Ïó∞ÏïàÌï®", 2: "ÎÅäÏùå", 3: "Ìù°Ïó∞Ï§ë"}
drink = {0: "ÏïàÎßàÏã¨", 1: "ÎßàÏã¨"}

# Ï≤¥Ï§ëÏù¥ 120Kg Ïù¥ÏÉÅÏù∏ Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
over_120kg = df[df['Ï≤¥Ï§ë(5Kg Îã®ÏúÑ)'] >= 120]

# Îß§Ìïë ÎîïÏÖîÎÑàÎ¶¨Î•º Ïù¥Ïö©ÌïòÏó¨ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Í∞í Î≥ÄÍ≤Ω
over_120kg["Ìù°Ïó∞ÏÉÅÌÉú"] = over_120kg["Ìù°Ïó∞ÏÉÅÌÉú"].map(smoke)
over_120kg["ÏùåÏ£ºÏó¨Î∂Ä"] = over_120kg["ÏùåÏ£ºÏó¨Î∂Ä"].map(drink)

# FacetGridÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
# sns.regplot ÏùÑ Ï∂îÍ∞ÄÌïòÎ©¥ ÌöåÍ∑ÄÏÑ†Ïù¥ Í∏∞Î°ùÎêúÎã§ tip
g = sns.FacetGrid(over_120kg, col="Ìù°Ïó∞ÏÉÅÌÉú", hue="ÏùåÏ£ºÏó¨Î∂Ä")
g.map(sns.regplot, "Ï¥ùÏΩúÎ†àÏä§ÌÖåÎ°§", "Í∞êÎßàÏßÄÌã∞Ìîº", scatter_kws={'s':10}, ci=None)

# Í∑∏ÎûòÌîÑ Ï†úÎ™©Í≥º Ï∂ï Î†àÏù¥Î∏î Ï∂îÍ∞Ä
g.set_axis_labels("Ï¥ùÏΩúÎ†àÏä§ÌÖåÎ°§", "Í∞êÎßàÏßÄÌã∞Ìîº")

g.add_legend(title="ÏùåÏ£ºÏó¨Î∂Ä")
plt.show()
````

````python
age_code = {1: '0~4ÏÑ∏',
 2: '5~9ÏÑ∏',
 3: '10~14ÏÑ∏',
 4: '15~19ÏÑ∏',
 5: '20~24ÏÑ∏',
 6: '25~29ÏÑ∏',
 7: '30~34ÏÑ∏',
 8: '35~39ÏÑ∏',
 9: '40~44ÏÑ∏',
 10: '45~49ÏÑ∏',
 11: '50~54ÏÑ∏',
 12: '55~59ÏÑ∏',
 13: '60~64ÏÑ∏',
 14: '65~69ÏÑ∏',
 15: '70~74ÏÑ∏',
 16: '75~79ÏÑ∏',
 17: '80~84ÏÑ∏',
 18: '85ÏÑ∏+'}
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/NHIS_OPEN_GJ_2017.CSV.zip", encoding="cp949")

# Ïã§Î™Ö Îç∞Ïù¥ÌÑ∞ Ï†úÍ±∞
df_filtered = df[df['ÏãúÎ†•(Ï¢å)'] != 9.9]
df_filtered = df_filtered[df_filtered['ÏãúÎ†•(Ïö∞)'] != 9.9]

df_filtered['Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)'] = df_filtered['Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)'].map(age_code)

# Ïó∞Î†πÎåÄÏôÄ ÏÑ±Î≥ÑÏóê Îî∞Î•∏ Ï¢å, Ïö∞ ÌèâÍ∑† ÏãúÎ†• Í≥ÑÏÇ∞
grouped = df_filtered.groupby(['Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)', 'ÏÑ±Î≥ÑÏΩîÎìú']).agg({'ÏãúÎ†•(Ï¢å)': 'mean', 'ÏãúÎ†•(Ïö∞)': 'mean'}).reset_index()

# ÏÑúÎ∏åÌîåÎ°Ø ÏÉùÏÑ±
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

# Ï¢å ÏãúÎ†•
sns.barplot(data=grouped, x='ÏãúÎ†•(Ï¢å)', y='Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)', hue='ÏÑ±Î≥ÑÏΩîÎìú', ax=ax[0])
ax[0].set_title("Ïó∞Î†πÎåÄ Î∞è ÏÑ±Î≥ÑÏóê Îî∞Î•∏ Ï¢å ÏãúÎ†• ÌèâÍ∑†")
ax[0].set_xlabel("ÌèâÍ∑† Ï¢å ÏãúÎ†•")
ax[0].legend(title="ÏÑ±Î≥Ñ")

# Ïö∞ ÏãúÎ†•
sns.barplot(data=grouped, x='ÏãúÎ†•(Ïö∞)', y='Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)', hue='ÏÑ±Î≥ÑÏΩîÎìú', ax=ax[1])
ax[1].set_title("Ïó∞Î†πÎåÄ Î∞è ÏÑ±Î≥ÑÏóê Îî∞Î•∏ Ïö∞ ÏãúÎ†• ÌèâÍ∑†")
ax[1].set_xlabel("ÌèâÍ∑† Ïö∞ ÏãúÎ†•")
ax[1].legend(title="ÏÑ±Î≥Ñ")

plt.tight_layout()
plt.show()
````
