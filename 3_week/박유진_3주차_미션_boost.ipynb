**Q1. Ïó∞Î†πÎåÄÎ≥Ñ ÌóàÎ¶¨ÎëòÎ†àÏóê ÎåÄÌïú Í∏∞Ïà†ÌÜµÍ≥ÑÎ•º Íµ¨ÌïòÎ†§Í≥† Ìï©ÎãàÎã§. Îã§Ïùå Ï†úÍ≥µÎêòÎäî ÎîïÏÖîÎÑàÎ¶¨Î•º ÌÜµÌï¥ Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)Î•º "Ïó∞Î†πÎåÄ"Î°ú ÎßåÎì§Í≥† ÏïÑÎûòÏôÄ Í∞ôÏùÄ Í∏∞Ïà†ÌÜµÍ≥ÑÍ∞íÏùÑ Íµ¨Ìï¥Ï£ºÏÑ∏Ïöî!**

```python
age_code = {1: '0~4ÏÑ∏',
 2: '5~9ÏÑ∏',
 3: '10~14ÏÑ∏',
 4: '15~19ÏÑ∏',
 5: '20~24ÏÑ∏',
 6: '25~29ÏÑ∏',
 7: '30~34ÏÑ∏',
 8: '35~39ÏÑ∏',
 9: '40~44ÏÑ∏',
 10: '45~49ÏÑ∏',
 11: '50~54ÏÑ∏',
 12: '55~59ÏÑ∏',
 13: '60~64ÏÑ∏',
 14: '65~69ÏÑ∏',
 15: '70~74ÏÑ∏',
 16: '75~79ÏÑ∏',
 17: '80~84ÏÑ∏',
 18: '85ÏÑ∏+'}
```

```python
def map_age_group(age_code):
    age_code_map = {
        1: '0~4ÏÑ∏',
        2: '5~9ÏÑ∏',
        3: '10~14ÏÑ∏',
        4: '15~19ÏÑ∏',
        5: '20~24ÏÑ∏',
        6: '25~29ÏÑ∏',
        7: '30~34ÏÑ∏',
        8: '35~39ÏÑ∏',
        9: '40~44ÏÑ∏',
        10: '45~49ÏÑ∏',
        11: '50~54ÏÑ∏',
        12: '55~59ÏÑ∏',
        13: '60~64ÏÑ∏',
        14: '65~69ÏÑ∏',
        15: '70~74ÏÑ∏',
        16: '75~79ÏÑ∏',
        17: '80~84ÏÑ∏',
        18: '85ÏÑ∏+',
    }
    return age_code_map.get(age_code, 'Ïïå Ïàò ÏóÜÏùå')
```

```python
df['Ïó∞Î†πÎåÄ'] = df['Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)'].apply(map_age_group)
```

```python
df.groupby('Ïó∞Î†πÎåÄ')['ÌóàÎ¶¨ÎëòÎ†à'].describe()
```

```python
df['Ïó∞Î†πÎåÄ'] = df['Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)'].replace(age_code)
df.groupby(['Ïó∞Î†πÎåÄ'])['ÌóàÎ¶¨ÎëòÎ†à'].describe()
```

ÏÑúÎ∏åÌîåÎ°ØÏúºÎ°ú ÏãúÍ∞ÅÌôîÌïòÏó¨ ÌëúÌòÑÌï¥ Ï£ºÍ≥† matplotlib Ïóê ÎåÄÌïú ÏÇ¨Ïö©Î≤ïÏùÄ Îã§ÏùåÏùò ÏòÅÏÉÅÎèÑ Ìï®Íªò Ï∞∏Í≥†Ìï¥ Ï£ºÏÑ∏Ïöî! (https://www.youtube.com/watch?v=ZTRKojTLE8M)

ÏãúÍ∞ÅÌôî ÏòàÏãúÎäî Ï∞∏Í≥†ÏÇ¨Ìï≠ÏúºÎ°ú Ïä§ÌÉÄÏùºÏù¥ÎÇò Ìè∞Ìä∏ Íµ¨ÏÑ±Ïù¥ Îã¨ÎùºÎèÑ Ïó∞Î†πÎåÄ, ÏÑ±Î≥Ñ Ï¢åÏö∞ÌèâÍ∑†ÏãúÎ†•ÏùÑ ÏãúÍ∞ÅÌôî ÌïòÎ©¥ Îê©ÎãàÎã§.

```python
# Ïã§Î™Ö Îç∞Ïù¥ÌÑ∞Î•º Ï†úÍ±∞
df_eye=df[df['ÏãúÎ†•(Ï¢å)'] != 9.9]
df_eye=df_eye[df_eye['ÏãúÎ†•(Ï¢å)'] != 9.9]

df_eye['Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)'] = df_eye['Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)'].map(age_code)

df_eye_group = df_eye.groupby(['Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)', 'ÏÑ±Î≥ÑÏΩîÎìú']).agg({'ÏãúÎ†•(Ï¢å)': 'mean', 'ÏãúÎ†•(Ïö∞)': 'mean'}).reset_index()
fig, ax = plt.subplots(1,2, figsize=(15,10))
sns.barplot(data=df_eye_group, x='ÏãúÎ†•(Ï¢å)', y='Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)', hue='ÏÑ±Î≥ÑÏΩîÎìú', ax=ax[0])
ax[0].legend(title='ÏÑ±Î≥Ñ')
sns.barplot(data=df_eye_group, x='ÏãúÎ†•(Ïö∞)', y='Ïó∞Î†πÎåÄÏΩîÎìú(5ÏÑ∏Îã®ÏúÑ)', hue='ÏÑ±Î≥ÑÏΩîÎìú', ax=ax[1])
ax[1].legend(title='ÏÑ±Î≥Ñ')
```
