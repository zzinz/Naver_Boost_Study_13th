**Q1. 연령대별 허리둘레에 대한 기술통계를 구하려고 합니다. 다음 제공되는 딕셔너리를 통해 연령대코드(5세단위)를 "연령대"로 만들고 아래와 같은 기술통계값을 구해주세요!**

```python
age_code = {1: '0~4세',
 2: '5~9세',
 3: '10~14세',
 4: '15~19세',
 5: '20~24세',
 6: '25~29세',
 7: '30~34세',
 8: '35~39세',
 9: '40~44세',
 10: '45~49세',
 11: '50~54세',
 12: '55~59세',
 13: '60~64세',
 14: '65~69세',
 15: '70~74세',
 16: '75~79세',
 17: '80~84세',
 18: '85세+'}
```

```python
def map_age_group(age_code):
    age_code_map = {
        1: '0~4세',
        2: '5~9세',
        3: '10~14세',
        4: '15~19세',
        5: '20~24세',
        6: '25~29세',
        7: '30~34세',
        8: '35~39세',
        9: '40~44세',
        10: '45~49세',
        11: '50~54세',
        12: '55~59세',
        13: '60~64세',
        14: '65~69세',
        15: '70~74세',
        16: '75~79세',
        17: '80~84세',
        18: '85세+',
    }
    return age_code_map.get(age_code, '알 수 없음')
```

```python
df['연령대'] = df['연령대코드(5세단위)'].apply(map_age_group)
```

```python
df.groupby('연령대')['허리둘레'].describe()
```

```python
df['연령대'] = df['연령대코드(5세단위)'].replace(age_code)
df.groupby(['연령대'])['허리둘레'].describe()
```

체중이 120Kg 이 넘는다면 비만으로 건강관리가 필요합니다. 120Kg 이상인 데이터를 찾아 "총콜레스테롤", "감마지티피" 값을 음주여부에 따라 산점도로 시각화해주세요! (이때 120Kg 도 포함되게 구합니다.) 그래프의 설정과 seaborn 버전에 따라 여백이나 회귀선의 시작과 끝점에 차이가 있을 수 있습니다.

📌 Q5. 연령대별로 시력은 얼마나 차이가 날까요? 연령대, 성별 좌우 평균 시력을 시각화 해주세요!

시력은 0.1~2.5 사이의 값으로 표기하며 0.1 이하의 시력은 0.1, 실명은 9.9로 표기합니다.

실명인 데이터를 평균에 포함하게 되면 시력에 대한 평균을 제대로 구하기 어렵습니다. 실명 데이터를 제외하고 연령대, 성별에 따른 평균 좌우 시력을 구해주세요!

서브플롯으로 시각화하여 표현해 주고 matplotlib 에 대한 사용법은 다음의 영상도 함께 참고해 주세요! (https://www.youtube.com/watch?v=ZTRKojTLE8M)

시각화 예시는 참고사항으로 스타일이나 폰트 구성이 달라도 연령대, 성별 좌우평균시력을 시각화 하면 됩니다.

```python
# 실명 데이터를 제거
df_eye=df[df['시력(좌)'] != 9.9]
df_eye=df_eye[df_eye['시력(좌)'] != 9.9]

df_eye['연령대코드(5세단위)'] = df_eye['연령대코드(5세단위)'].map(age_code)

df_eye_group = df_eye.groupby(['연령대코드(5세단위)', '성별코드']).agg({'시력(좌)': 'mean', '시력(우)': 'mean'}).reset_index()
fig, ax = plt.subplots(1,2, figsize=(15,10))
sns.barplot(data=df_eye_group, x='시력(좌)', y='연령대코드(5세단위)', hue='성별코드', ax=ax[0])
ax[0].legend(title='성별')
sns.barplot(data=df_eye_group, x='시력(우)', y='연령대코드(5세단위)', hue='성별코드', ax=ax[1])
ax[1].legend(title='성별')
```
